## Visualizar Métricas de Performance



```python
import matplotlib.pyplot as plt

# Create the bar chart
plt.figure(figsize=(8, 6))
plt.bar(metrics_names, metrics_values, color=['skyblue', 'lightcoral', 'lightgreen', 'gold'])

# Add title and labels
plt.title('Métricas de Performance do Modelo')
plt.xlabel('Métrica')
plt.ylabel('Valor')
plt.ylim(0, 1.0) # Assuming metrics are between 0 and 1

# Display the values on top of the bars
for i, value in enumerate(metrics_values):
    plt.text(i, value + 0.02, f'{value:.2f}', ha='center', va='bottom')

# Show the plot
plt.show()
```

## Visualizar Distribuição das Classes Reais


```python
import matplotlib.pyplot as plt

# Calculate the count of each real class
class_distribution = df_results['real_class'].value_counts()

# Create the bar chart
plt.figure(figsize=(7, 5))
class_distribution.plot(kind='bar', color=['lightseagreen', 'indianred'])

# Add title and labels
plt.title('Distribuição das Classes Reais')
plt.xlabel('Classe Real')
plt.ylabel('Contagem')
plt.xticks(rotation=0) # Ensure labels are horizontal

# Display the values on top of the bars
for i, count in enumerate(class_distribution):
    plt.text(i, count + 0.5, str(count), ha='center', va='bottom')

# Show the plot
plt.show()
```

## Visualizar Confiança em Predições Incorretas



```python
import matplotlib.pyplot as plt

# Create a figure and a set of axes
plt.figure(figsize=(10, 6))

# Generate the first histogram for false positives
plt.hist(
    false_positives_df['prob_malign'],
    bins=10,  # You can adjust the number of bins
    alpha=0.7,
    label='Falsos Positivos: Prob. Maligno',
    color='orange',
    edgecolor='black'
)

# Generate the second histogram for false negatives
plt.hist(
    false_negatives_df['prob_benign'],
    bins=10,  # Use the same number of bins for consistency
    alpha=0.7,
    label='Falsos Negativos: Prob. Benigno',
    color='purple',
    edgecolor='black'
)

# Add title and labels
plt.title('Distribuição das Probabilidades para Predições Incorretas')
plt.xlabel('Probabilidade')
plt.ylabel('Contagem')

# Add a legend to differentiate the two histograms
plt.legend()

# Set x-axis limits for probability (0 to 1)
plt.xlim(0, 1)

# Display the plot
plt.show()
```
